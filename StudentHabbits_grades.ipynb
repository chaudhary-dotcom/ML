```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, accuracy_score
```

```python
# Model 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
df.head()
```

```python
categorical_cols = ['gender', 'part_time_job', 'diet_quality', 'parental_education_level', 'internet_quality','extracurricular_participation']
le = LabelEncoder()

for cols in categorical_cols:
    df[cols] = le.fit_transform(df[cols])

df.head()
```

```python
X = df.drop(['student_id', 'exam_score'], axis=1)

def assign_grade(score):
    if 80 <= score <= 100:
        return 'A Grade'
    
    elif 60 <= score < 80:
        return 'B Grade'
    
    elif 40 <= score <60:
        return 'C grade'
    
    else: 
        return 'Fail'

y = [assign_grade(score) for score in df['exam_score']]

y = le.fit_transform(y)

# train test split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=42)
```

```python
models = {
    'LogisticRegression': LogisticRegression(),
    'decision Tree': DecisionTreeClassifier(),
    'k Neighbours': KNeighborsClassifier()
    
}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)

    test_acc = accuracy_score(y_test, pred_test) * 100
    train_acc = accuracy_score(pred_train, y_train) * 100

    train_cm = confusion_matrix(y_train, pred_train)
    test_cm = confusion_matrix(y_test, pred_test)

    print(f'{name} Train accuracy: {train_acc:.2f}')

    dispp = ConfusionMatrixDisplay(confusion_matrix=train_cm)
    dispp.plot(cmap='Blues')
    plt.title(f'confusion matri - {name}')
    plt.show()
    
    print(f'{name} Test accuracy: {test_acc:.2f}')

    dispp = ConfusionMatrixDisplay(confusion_matrix=test_cm)
    dispp.plot(cmap='Blues')
    plt.title(f'confusion matri - {name}')
    plt.show()
```

```python
params = {
    'penalty': ['l2','l1','elasticnet'],
    'solver': ['lbfgs','liblinear','newton-cg','newton-cholesky','sag','saga'],
    'max_iter': [100,500,1000],   
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
lr = LogisticRegression()
clf = GridSearchCV(param_grid=params, estimator=lr)
clf.fit(X_train, y_train)
```

```python
print(clf.best_params_)
```

```python
print(clf.best_score_)
```

```python
test_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, test_pred)
print(accuracy)
```
